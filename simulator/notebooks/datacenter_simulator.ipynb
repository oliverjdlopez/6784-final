```python
plt.plot(sim_no_pre_timestep_rewards["num_agent_steps_sampled"], sim_no_pre_timestep_rewards["episode_reward_mean"], label="Without Pre-Training")

plt.title("Agent reward vs. Timestep")
plt.xlabel("Timestep")
plt.ylabel("Reward")
plt.legend()
plt.show()
```
